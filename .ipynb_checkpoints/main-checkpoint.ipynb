```python
from Independent_vars import Enverionmental_Factor
from Thermal_Model import Thermal_Model
import matplotlib.pyplot as plt
```

```python
env_1 = Enverionmental_Factor(met = 58.2, h_r = 5.23, h_c = 2.91, clo = 0.6, db_T = 30, phi = 0.5, mu = 0)
model_1 = Thermal_Model(env_Factor = env_1)
t = 0
t_list = []
T_cr = []
T_sk = []
while t < 1:
    dt = model_1.heat_balance()
    model_1.self_control(dt)
    t = t + dt
    t_list.append(t)
    T_cr.append(model_1.T_cr)
    T_sk.append(model_1.T_sk)

plt.plot(t_list, T_cr)  #plot core temperature
plt.ylim((34, 40))
plt.show()
```
