```python
from Independent_vars import Enverionmental_Factor
from Thermal_Model import Thermal_Model
import matplotlib.pyplot as plt
```

```python
env_1 = Enverionmental_Factor(met = 58.2, h_r = 5.23, h_c = 2.91, clo = 0.6, db_T = 30, phi = 0.5, mu = 0)
model_1 = Thermal_Model(env_Factor = env_1)
t = 0
t_list = []
T_cr = []
T_sk = []
while t < 1:
    dt = model_1.heat_balance()
    model_1.self_control(dt)
    t = t + dt
    t_list.append(t)
    T_cr.append(model_1.T_cr)
    T_sk.append(model_1.T_sk)

plt.figure(1)
plt.plot(t_list, T_cr)  #plot core temperature
plt.ylim((34, 40))
plt.show()
```

```python
T_cr = []
T_sk = []
V_bl = []
E_rsw = []
for phi in [0.3, 0.5, 0.9]:
    T_cr.append([])
    T_sk.append([])
    V_bl.append([])
    E_rsw.append([])
    for T in range(5, 46):
        #env = Enverionmental_Factor(met = 130.1, h_r = 5.23, h_c = 4.3, clo = 0.6, db_T = T, phi = phi, mu = 0)
        env = Enverionmental_Factor(met = 58.2, h_r = 5.23, h_c = 2.91, clo = 0.6, db_T = T, phi = phi, mu = 0)
        model = Thermal_Model(env_Factor = env)
        t = 0
        while t < 1:
            dt = model.heat_balance()
            model.self_control(dt)
            t = t + dt
        T_cr[-1].append(model.T_cr)
        T_sk[-1].append(model.T_sk)
        V_bl[-1].append(model.V_bl)
        E_rsw[-1].append(model.E_rsw)

T = list(range(5,46))
phi = ['0.3', '0.5', '0.9']
plt.figure(2)
plt.ylim((25,40))
for i in range(3):
    plt.plot(T, T_sk[i], label = "T_sk "+phi[i])
    plt.legend()
plt.show


plt.figure(3)
#plt.ylim((30,40))
for i in range(3):
    plt.plot(T, T_cr[i], label = "T_cr "+phi[i])
    plt.legend()
plt.show

plt.figure(4)
#plt.ylim((30,40))
for i in range(3):
    plt.plot(T, V_bl[i], label = "V_bl "+phi[i])
    plt.legend()
plt.show

plt.figure(5)
#plt.ylim((30,40))
for i in range(3):
    plt.plot(T, E_rsw[i], label = "E_rsw RH="+phi[i])
    plt.legend()
plt.show
```
